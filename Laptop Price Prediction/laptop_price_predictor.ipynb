```python
data=data.drop(['Unnamed: 0','TypeName','OpSys','Weight'],axis='columns')
```

```python
data["ScreenResolution"].value_counts()
```

```python
data['display'] = data["ScreenResolution"].apply(lambda x: 'ips' if 'IPS Panel' in x else x)
```

```python
data['display'] = data['display'].apply(lambda x: 'full hd' if 'Full HD' in x else x)
```

```python
data['display'] = data['display'].apply(lambda x: 'touch screen' if 'Touchscreen' in x else x)
```

```python
data['display'].value_counts()
```

```python
def screen(display):
    if display == 'full hd' or display == 'ips' or display == 'touch screen':
        return display
    else:
        return 'other'
```

```python
data['display'] = data['display'].apply(screen)
```

```python
data=data.drop('ScreenResolution',axis='columns')
```

```python
data.head()
```

```python
data.head()
```

```python
data['Cpu'].value_counts()
```

```python
data['processor'] = data['Cpu'].apply(lambda x: 'i5' if  'Intel Core i5' in x else x)
```

```python
data['processor'] = data['processor'].apply(lambda x: 'i7' if  'Intel Core i7' in x else x)
```

```python
data['processor'] = data['processor'].apply(lambda x: 'i3' if  'Intel Core i3' in x else x)
```

```python
data['processor'] = data['processor'].apply(lambda x: 'amd processor' if  'AMD' in x else x)
```

```python
data['processor'] = data['processor'].apply(lambda x: 'other intel processor' if  'Intel' in x else x)
```

```python
data['processor'].value_counts()
```

```python
data=data.drop('Cpu',axis='columns')
```

```python
data.head()
```

```python
data['Ram'] = data['Ram'].str.replace('GB','')
```

```python
data['Ram'] = data['Ram'].astype(int)
```

```python
data.head()
```

```python
data["Memory"] = data["Memory"].str.replace("GB","")
```

```python
data["Memory"] = data["Memory"].str.replace("TB","000")
```

```python
data.head(30)
```

```python
new = data['Memory'].str.split("+", n = 1, expand = True)
```

```python
new.head(30)
```

```python
data['layer_1_hdd'] = new[0]
```

```python
data['layer_1_hdd'] = data['layer_1_hdd'].apply(lambda x: 1 if 'HDD' in x else 0)
```

```python
data['layer_2_hdd'] = new[1].fillna('0')
```

```python
data['layer_2_hdd'] = data['layer_2_hdd'].apply(lambda x: 1 if 'HDD' in x else 0)
```

```python
data['first'] = new[0]
data['first'] = data['first'].str.strip()
```

```python
data['first'] = data['first'].str.replace(r'\D','')
```

```python
data['secound'] = new[1].fillna('0')
data['secound'] = data['secound'].str.strip()
```

```python
data['secound'] = data['secound'].str.replace(r'\D','')
```

```python
data['secound'] = data['secound'].astype(int)
data['first'] = data['first'].astype(int)
```

```python
data['HDD'] = (data['first'] * data['layer_1_hdd'] + data['secound'] * data['layer_2_hdd'])
```

```python
data.head(40)
```

```python
data['layer_1_ssd'] = new[0]
data['layer_1_ssd'] = data['layer_1_ssd'].apply(lambda x: 1 if 'SSD' in x else 0)
```

```python
data['layer_2_ssd'] = new[1].fillna('0')
data['layer_2_ssd'] = data['layer_2_ssd'].apply(lambda x: 1 if 'SSD' in x else 0)
```

```python
data['SSD'] = (data['first'] * data['layer_1_ssd'] + data['secound'] * data['layer_2_ssd'])
```

```python
data.head(30)
```

```python
data = data.drop(['layer_1_hdd','layer_2_hdd','first','secound','layer_1_ssd','layer_1_ssd'],axis='columns')
```

```python
data = data.drop(['layer_2_ssd'],axis='columns')
```

```python
data.head(30)
```

```python
graphic = data['Gpu'].value_counts()
```

```python
graphic_less_then_10 = graphic[graphic<=10]
```

```python
data['Gpu'] = data['Gpu'].apply(lambda x: 'other' if x in graphic_less_then_10 else x)
```

```python
data.head(30)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import pickle
```

```python
data= data.drop('Memory',axis='columns')
```

```python
X = data.drop('Price', axis='columns')
Y = data.Price
```

```python
data.head()
```

```python
data.to_csv('new_data.csv')
```

```python
ohe= OneHotEncoder(sparse=False,drop='first')
```

```python
ohe.fit(data[['Company','Gpu','display','processor']])
```

```python
X_test, X_train, y_test, y_train = train_test_split(X,Y,train_size=0.2)
```

```python
transform = make_column_transformer((OneHotEncoder(categories=ohe.categories_,sparse=False,drop='first'),['Company','Gpu','display','processor']),remainder='passthrough')
```

Liner Regression


```python
lr = LinearRegression()
pipe = make_pipeline(transform,lr)
pipe.fit(X_train,y_train)
prediction = pipe.predict(X_test)
r2_score(y_test,prediction)
```

Random Forest


```python
rr = RandomForestRegressor()
pipe = make_pipeline(transform,rr)
pipe.fit(X_train,y_train)
prediction = pipe.predict(X_test)
r2_score(y_test,prediction)
```

```python
pickle.dump(pipe,open('model.pkl','wb'))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso
```

Decision Tree

```python
dr = DecisionTreeRegressor()
pipe = make_pipeline(transform,dr)
pipe.fit(X_train,y_train)
prediction = pipe.predict(X_test)
r2_score(y_test,prediction) 
```

Lasso

```python
lasso = Lasso()
pipe = make_pipeline(transform,lasso)
pipe.fit(X_train,y_train)
prediction = pipe.predict(X_test)
r2_score(y_test,prediction)
```
